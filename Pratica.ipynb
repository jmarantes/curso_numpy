```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
dado
```

```python
dim_laranja = dado[:5000, 0] 
dim_toranja = dado[5000:, 0]
peso_laranja = dado[:5000, 1]
peso_toranja = dado[5000:, 1]
```

```python
plt.plot(dim_laranja, peso_laranja)
plt.plot(dim_toranja, peso_toranja)
plt.legend(['Laranja', 'Toranja'])
plt.show()
```

```python
YL = peso_laranja
XL = dim_laranja
nL = np.size(dim_laranja)
```

```python
aL = (nL * np.sum(XL * YL) - np.sum(XL) * np.sum(YL)) / (nL * np.sum(XL ** 2) - np.sum(XL) ** 2)
aL
```

```python
bL = np.mean(YL) - aL * np.mean(XL)
bL
```

```python
yL = aL * XL + bL
plt.plot(XL, yL)
plt.plot(dim_laranja, peso_laranja)
```

```python
YT = peso_toranja
XT = dim_toranja
nT = np.size(dim_toranja)
```

```python
aT = (nT * np.sum(XT * YT) - np.sum(XT) * np.sum(YT)) / (nT * np.sum(XT ** 2) - np.sum(XT) ** 2)
aT
```

```python
bT = np.mean(YT) - aT * np.mean(XT)
bT
```

```python
yT = aT * XT + bT
plt.plot(XT, yT)
plt.plot(dim_toranja, peso_toranja)
```

```python
b = 17
```

```python
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.0, high=30.0, size=100)

norma = np.array([])
for i in range(100):
    norma = np.append(norma, np.linalg.norm(peso_laranja - (coef_angulares[i] * XT + b)))
    if norma[i] <= min(norma):
        min_norma = norma[i]
        i_min = i

print(f'Menor norma laranja: {min_norma}')
print(f'Coeficiente angular laranja = {coef_angulares[i_min]}')
```
